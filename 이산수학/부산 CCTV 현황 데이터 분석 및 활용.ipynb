**용어 정리**

**데이터프레임(Data Frame)**: 행과 열로 이루어진 표로써 데이터분석을 위한 수집, 전처리 등의 과정에서 사용된다.

**시리즈(Series)**: 데이터프레임의 하위 자료형으로, 1개의 **열**이 시리즈이고 이 시리즈가 다수 모여 데이터프레임을 형성한다.


````python
# 시리즈(Series) 생성
''' 
dictionary(딕셔너리)와 list(리스트)를 각각 만들어 보겠습니다
'''
dict_data = {'a':1, 'b':2, 'c': 3}
list_data = ['2021-09-24', 3.14, 'ABC', 100, True]
````

````python
dict_data
````

````python
list_data
````

그럼 각각의 dictionary(딕셔너리)와 list(리스트)를 **시리즈(열)**로 바꿔보겠습니다. 

````python
dict_data = pd.Series(dict_data)
dict_data
````

````python
list_data = pd.Series(list_data)
list_data
````

이렇게 2가지 열이 생성되었습니다. **뭔가 둘이 다른점을 느끼셨나요??**

이번엔 데이터프레임을 만들어보겠습니다.

````python
# dictionary로 데이터프레임 만들기
dict_data = {'c0':[1,2,3],'c1':[4,5,6],'c2':[7,8,9],'c3':[10,11,12],'c4':[13,14,15]}
dict_data = pd.DataFrame(dict_data)
dict_data
````

````python
# list로 데이터프레임 만들기
# list는 행과 열의 요소 이름을 지정하기 쉽다.
list_data = pd.DataFrame([[14,'남','염경중'],[17,'여','양동중']],
                  index = ['준서','예은'],
                  columns = ['나이','성별','학교'])
list_data
````

이제는 차이점이 보이시나요?? 

dictionary는 **열(column)**을 하나씩 쌓아가는 형태이고,

## **Chapter 2 - 데이터프레임 다루기**

---



**행, 열 이름 변경: rename()함수**

````python
list_data = pd.DataFrame([[14,'남','염경중'],[17,'여','양동중']],
                  index = ['준서','예은'],
                  columns = ['나이','성별','학교'])
list_data
````

rename( )함수는 데이터프레임 자료형에서 제공하는 함수이다.

df.rename(columns={변경 전 열이름, 변경 후 열이름}, inplace=True)로 사용할 수 있고 

inplace=True옵션을 지정해주면 df에 바로 적용된다.

````python
list_data.rename(columns={'나이':'연령','성별':'남녀', '학교':'소속'}, inplace=True)
list_data
````

````python
list_data.rename(index={'준서':'학생1', '예은':'학생2'}, inplace=True)
list_data
````



---
**행, 열 삭제: drop()함수**

drop('행 또는 열이름', axis = 0 or 1, inplace =True)와 같이 사용한다.
행을 삭제하려면 axis=0(디폴트값), 열을 삭제하려면 axis=1을 꼭 지정해 주어야한다.


````python
exam_data = {'수학':[90,80,70], '영어':[98,89,95],
            '음악':[85,95,100], '체육':[100,90,90]}
df = pd.DataFrame(exam_data, index = ['학생1','학생2','학생3'])
df
````

````python
df2 = df.copy() # df에 영향을 주지 않기 위해 df내용을 담고있는 df2 생성
df2.drop('학생2',inplace=True)  #행삭제는 axis옵션 생략가능
df2
````

````python
df3 = df.copy()
df3.drop(['수학','음악'],axis=1,inplace=True)
df3
````



---
**행, 열 추가**

열추가는 df['새로운 열이름'] = 값으로 쉽게 추가해줄 수 있다.

물론 새로운 열이름이 아니라 기존 열이름이라면 덮어씌워진다.

````python
# 데이터프레임 생성
exam_data = {'이름':['지원','민혁','건우']
             ,'수학':[90,80,90], '영어':[98,89,100],
            '음악':[85,95,100], '체육':[10,90,90]}
df = pd.DataFrame(exam_data)
df['국어'] = 80
df
````

````python
# 행추가는 loc(인덱스명)와 iloc(위치)를 활용해서 쓸수있는데 자세한 내용은 필터링에서~
exam_data = {'이름':['지원','민혁','건우']
             ,'수학':[90,80,90], '영어':[98,89,100],
            '음악':[85,95,100], '체육':[10,90,90]}
df = pd.DataFrame(exam_data)
df.loc[3] = 10
df.loc[4] = ['용근',100,100,100,100]
df
````

이처럼 행추가를 할때 단일값은 그 값으로 모두 들어가고, 그게 아니라면 행또는 열의 수에 맞게 원소를 추가시켜야 한다.

## **Chapter 3 - 필터링**

---
데이터를 전처리하기위해서 행과 열의 필터링이 필요하다.


**행선택(인덱싱): loc와 iloc**


**loc**와 **iloc**는 필터링 할때 많이 사용하는 함수이다. 

**loc**는 인덱스명으로, **iloc**는 숫자(위치)로 인덱싱한다고 생각하고 아래 내용을 봐보자.

````python
# 새로운 데이터프레임 생성
exam_data = {'수학':[90,80,100], '영어':[98,89,100],
            '음악':[85,95,100], '체육':[100,60,100]}
df = pd.DataFrame(exam_data, index = ['재혁','지운','세진'])
df
````

첫번째 행('세진')을 선택해보자

````python
df_index1 = df.loc['세진']
df_position1 = df.iloc[2]
````

````python
df_index1
````

````python
df_position1
````

결과값이 마치 열을 선택한 것처럼 나와서 처음보면 헷갈릴 수도 있다. 

하지만 재혁이라는 행을 **인덱스명**과 **행위치(index)**로 선택한 것을 알 수 있다.







---


다음으로 두개 이상의 행을 선택해볼텐데 이때는 대괄호를 신경써야한다

각각 선택은 두개 [[ ]], 범위( : )로 

한개는 위에서 한 것처럼 [ ]이다.

````python
df_index2 = df.loc[['재혁', '지운']]
df_position2 = df.iloc[[0,1]]
````

````python
df_index2
````

````python
df_position2
````

## ■ 문제 3-6. isSubset 함수 만들기
만약 데이터프레임 dfA, dfB가 주어질 때,
dfA가 dfB의 서브셋이고, dfB가 dfA의 서브셋이라면(동치=equivalence) return True
만약 dfA가 dfB의 서브셋이고 그 반대는 성립하지 않는 경우에도 return True
그 외(서로의 서브셋이 아닌 경우, dfB가 dfA의 서브셋이지만 그 반대는 성립하지 않는 경우)는 return False
을 반환하는 함수를 작성하여라

````python
def isSubset(dfA, dfB):
  # 인덱싱 번호를 리스트로 전환
 A=dfA.index.tolist()
 B=dfB.index.tolist()  
  # 리스트 정렬
 A.sort()
 B.sort()
 if(len(A)!=len(B)): #A와 B의 갯수가 다르고
    if(len(A)<len(B)): #B의 갯수가 더 많다면
      com=list(filter(lambda x: x not in A, B))  #B에 있는 것 중에 A에 없는거 com에 저장
      com.sort() 
      if(com!=B): ##B와 com이 다르고
        if(A+com).sort()==B: #ex) A={1,2,3,} B={1,2,3,4} com={4} A+com={1,2,3,4}  dfA가 dfB의 서브셋이다.
          return True
        elif(com==B): ## 서로의 서브셋이 아닌 경우
          return False
      
    elif(len(A)>len(B)):
      com=list(filter(lambda x: x not in B, A)) 
      com.sort() 
      if(com!=A): 
        if(B+com).sort()==A: ##dfB가 dfA의 서브셋이지만 그 반대는 성립하지 않는 경우 
          return False
        elif(com==0): ##서로의 서브셋이 아닌 경우
          return False  
        else:
         return False       
 elif(A==B and len(A)==len(B)): ## 동치
    return True
 

isSubset(P, I)
isSubset(P, P)
isSubset(P, S)
isSubset(P, P_only)
````
